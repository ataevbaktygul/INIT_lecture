## Stat functions with NaN.

Все вышеописанные функции основаны так или иначе на мат вычислениях. Но любая арифметическая операция с nan возвращает nan. В случае наличия пропусков полезными бывают следующие аналоги:

* `np.nanmean(arr: ... axis = ...)`
* `np.nanmax(arr: ... axis = ...)`
* `np.nanmin(arr: ... axis = ...)`
* `np.nanmedian(arr: ... axis = ...)`
* `np.nansum(arr: ... axis = ...)`
* ....

```python
nan_array = np.array([[1, 2, 3], [4, np.nan, 6], [7, 8, 9]])
nan_array
```

```python
nan_array.mean(axis=1)
```

```python
np.nanmean(nan_array, axis=1)
```
